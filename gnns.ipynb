```python
from torch_geometric.datasets import KarateClub

dataset = KarateClub()

# dataset exploration
print(dataset)
print('\n')
print(f"number of graphs: {len(dataset)}")
print(f"number of features: {dataset.num_features}")
print(f"number of classes: {dataset.num_classes}")
print(f"Graph: {dataset[0]}")
print(f"x = {dataset[0].x.shape}")
print(dataset[0].x)
print(f"edge index = {dataset[0].edge_index.shape}")
print(dataset[0].edge_index)
```

1 graph, each node has a feature vector of 34 elements, i.e, R^34 and is part of 1 out of the 4 classes (which means 4 groups).

aother observation we can make is that the shape of the feature matrix represents an identity matrix, implying all the node attributes do not contain any relevant information. We will have to classify our nodes by looking at connections.
